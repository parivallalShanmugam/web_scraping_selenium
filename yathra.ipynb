```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://insights.blackcoffer.com/ml-and-ai-based-insurance-premium-model-to-predict-premium-to-be-charged-by-the-insurance-company/')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

 
wait = WebDriverWait(driver, 10)
text_div = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.td-post-content.tagdiv-type")))

# Extract and print the text from the div element
text_ = text_div.text
print(text_)

# Close the driver
driver.quit()
```

```python
word_list = text_.split()
```

```python
import re


# Function to remove special characters and spaces
def clean_word(word):
    return re.sub(r'\W+', '', word)

# Clean each word in the list
cleaned_words = [clean_word(word) for word in word_list]

print(cleaned_words)
```

```python
import re

# Original list of words

# Function to clean word by removing special characters and trailing spaces
def clean_word(word):
    
    word = re.sub(r'[^a-zA-Z0-9\s]', '', word)
    
    word = word.strip()
    return word

# Clean each word in the list
cleaned_words = [clean_word(word) for word in word_list]

# Filter out any empty strings that may result from cleaning
cleaned_words = [word for word in cleaned_words if word]

print(cleaned_words)
```
